```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
df=load_iris()
```

```python
x=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
x.head()
```

```python
y=pd.DataFrame(df.target,columns=['species'])
```

```python
print(df.DESCR)
```

```python
corr=x.corr()
corr
```

## Principle Component Analysis

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaledx=scaler.fit_transform(x)
```

```python
scaledx
```

```python
## Running PCA
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
```

```python
components=pca.fit_transform(scaledx)
```

```python
components
```

```python
pca_feature=pd.DataFrame(components,columns=['pc1','pc2','pc3'])
```

```python
pca_feature.head()
```

```python
pca_feature.corr()
```

```python
pca.explained_variance_ratio_
```

```python
x=pca_feature
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
# Implementing Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
dt_pred=dt.predict(xtest)
# checking the accuracy of Decision Tree model
adt=accuracy_score(ytest,dt_pred)
adt
```
